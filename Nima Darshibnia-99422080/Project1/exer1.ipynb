```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('immo_data.csv')
```

```python
#حذف داده های پرت
remove_outliers = df[df > (df.mean() - df.std())*3]

#مدیریت Null
#dropna() چون همه نمونه ها دارای حداقل یک فیچر خالی هستند این روش گزینه خوبی نیست
df.dropna()

#مدیریت Null
#حذف فیچرهایی که نسبت زیادی از نمونه ها فاقد این فیچر هستند
df.dropna(axis=1, thresh=int(0.5*len(df)))

#مدیریت Null
#پرکردن داده ها با پرتکرارترین
df.fillna(df.mode().iloc[0])

#مدیریت Null
#پرکردن داده ها با میانگین مقادیر غیر Null
df = df.fillna(df.mean())
```

```python
#انجام پیش پردازش بر روی داده ها
#تبدیل مقادیر دسته ای به مقادیر عددی
cat_col_list = list(df.loc[:, df.dtypes == np.object].astype('category'))
for i in cat_col_list:
    df[i] = df[i].astype('category')
cat_columns = df.select_dtypes(['category']).columns
df[cat_columns] = df[cat_columns].apply(lambda x: x.cat.codes)
```

```python
# سوال 1
#کدام فیچرها بیشترین کورولیشن خطی را با هدف دارا میباشند
corrdf = df.dropna(axis=1, thresh=int(0.5*len(df)))
corrdf = corrdf.fillna(df.mean())
corr = corrdf.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
#3 سوال
#تعریف یک مدل رگرسیون بدون استفاده از پکیج
df = df.dropna(axis=1, thresh=int(0.5*len(df)))
df = df.fillna(df.mean())

cat_col_list = list(df.loc[:, df.dtypes == np.object].astype('category'))
for i in cat_col_list:
    df[i] = df[i].astype('category')
cat_columns = df.select_dtypes(['category']).columns
df[cat_columns] = df[cat_columns].apply(lambda x: x.cat.codes)
```

```python
#استفاده از همه فیچرها در مدل رگرسیونی
#پاسخ سوال 4

features = df.to_numpy()
target = df['garden'].to_numpy() 
weights = np.random.rand(42)
b = np.random.rand(1) 
bias = np.array([b[0] for i in range(len(features))])

def linearRegr(features, weights, bias):
    y_hat = weights.dot(features.transpose()) + np.array([bias[0] for i in range(len(features))]) # Takes the value stored in the bias array and makes an array of length of feature matrix for addition
    return y_hat

y_hat = linearRegr(features, weights, b)

def meanSqrError(y, y_hat):
    MSE = np.sum((y - y_hat) ** 2) / len(y)
    return MSE
print('Total error- {}'.format(meanSqrError(target, y_hat)))

#Absolute Error
def MAPE(y, y_hat):
    mape = np.mean(np.abs((y - y_hat)/y_hat))*100
    return mape
print('MAPE- {}'.format(MAPE(target, y_hat)))
```

```python
#استفاده از همه فیچرها در مدل رگرسیونی
#با استفاده از پکیج آماده sklearn
#پاسخ سوال 5 و 6
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import accuracy_score


y = df['garden']
x = df[:]

Linear_Regression = LinearRegression()
Linear_Regression.fit(x,y)

y_pred = Linear_Regression.predict(x)

def meanSqrError(y, y_pred):
    MSE = np.sum((y - y_pred) ** 2) / len(y)
    return MSE
print('MSE- {}'.format(meanSqrError(y, y_pred)))

#Absolute Error
def MAPE(y, y_pred):
    mape = np.mean(np.abs((y - y_pred)/y_pred))
    return mape
print('MAE- {}'.format(MAPE(y, y_pred)))

sklearn_Reg_df = pd.DataFrame({'Actual': y, 'Predicted': y_pred})
for i, row in sklearn_Reg_df.iterrows():
  sklearn_Reg_df.loc[i,"Absolute Error"] = ((row["Actual"]-row['Predicted'])/row['Predicted'])
print('Mean of All MAE is: ' , sklearn_Reg_df['Absolute Error'].mean())

for i, row in sklearn_Reg_df.iterrows():
  sklearn_Reg_df.loc[i,"True labels accuracy"] = (1.1>row["Actual"]>0.9)
print(sklearn_Reg_df)
```
