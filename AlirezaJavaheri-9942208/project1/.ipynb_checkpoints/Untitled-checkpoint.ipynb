## Remove outliers 

```python
def cal_iqr(key, data = data):
    q25, q75 = np.nanpercentile(data[key], 25), np.nanpercentile(data[key], 75)
    iqr = q75 - q25
    print('Percentiles: 25th= %.3f, 75th= %.3f, IQR= %.3f' % (q25, q75, iqr))
    cut_off = iqr * 1.5
    lower, upper = q25 - cut_off, q75 + cut_off
    outliers = [x for x in data if x < lower or x > upper]
    print('outlier: {}'.format(len(outliers)))
    new_data = data.loc[(data[key].isnull()) | ((data[key] > lower) & (data[key] < upper))]
    return new_data
```

```python
def plot_item(key, data = data):
    fig, axes = plt.subplots(3, figsize = (8, 10)) 
    sns.boxplot(x=data[key], ax=axes[0])
    sns.violinplot(x=data[key], ax=axes[1])
    sns.distplot(data[key], ax=axes[2])
    plt.show()
    print(data[key].describe())
```
